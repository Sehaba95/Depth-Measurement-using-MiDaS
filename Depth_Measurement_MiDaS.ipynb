## **Get the frames of a video and save them in the input folder, to use them by the MiDas model**

```
video_path = "Put your video path here"

# Load the video file from the give path
cap = cv2.VideoCapture(video_path)

# Specify where to put the images
input_path = 'input/'

# Open the images dataset
frames_counter = 0 

while(cap.isOpened()):
  # Get the frame from the video
  ret, frame = cap.read()
  if not ret:
    break
  # Prepare the output image filename
  image_name = input_path+'input_'+str(frames_counter)+'.png'
  # Write the name into the dataset txt file
  # Save the image in the specified path
  cv2.imwrite(image_name,frame)
  frames_counter += 1 

cap.release()
cv2.destroyAllWindows()
print("Number of images: {} ".format(frames_counter))
```

### **Create a depth video using the large model**

```
import glob
import cv2


images_path = "/content/drive/My Drive/Colab Notebooks/AINA/MiDaS/output/"
img_array = []

filenames = []

for filename in glob.glob(images_path+'*.png'):
  filenames.append(filename)
  print(filename)

for index in range(0,len(filenames)):
  # This works only for the MiDaS solution
  filename = '/content/drive/My Drive/Colab Notebooks/AINA/MiDaS/output/input_'+str(index)+'.png'
  img = cv2.imread(filename)
  height, width, layers = img.shape
  size = (width,height)
  img_array.append(img)
  
out = cv2.VideoWriter('MiDaS_output_large_model.mp4',cv2.VideoWriter_fourcc(*'mp4v'), 20, size)
for i in range(len(img_array)):
  out.write(img_array[i])
out.release()
print("Depth Video Created Successfully.")
```
